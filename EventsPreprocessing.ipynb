## Data Preprocessing 

From the above dataframe we can derive:
 **Mode** looks like an integer value but it is actually a categorical variable;

The dataset has missing values (shown with `"?"`s). The device had a problem so parts of the readings from **Mode** and **Sensor 3** was lost. This loss is **MCAR (Missing Completely At Random)**;

Values of **Sensor 9** missing are actually due to a faulty sensor that might have not recorded readings in certain conditions, and expert tells you that the faulty sensor may happen when readings of **Sensor 7** have values less than a certain threshold;

All readings from **Sensor 6** are showing the same value;

The different feature values come in different ranges;

Also, we will want to remove rows whose value for column named `"Events"` is `"-"`.

### 1. Removing Irrelevant class (datapoints)

```python
is_false_detection = (data["Events"] == "-")
```

```python
display(is_false_detection[:10].values)
print("Total number of false detections:", is_false_detection.values.astype('uint').sum())
```

```python
false_detection_row_indices = data[is_false_detection].index
```

```python
display(false_detection_row_indices[:10].values)
```

```python
data = data.drop(index=false_detection_row_indices)
```

```python
data
```

We can see the first 10 labels, which do not include any `"-"`s in the output.


### 2. Removing constant-value feature

Removing single valued feature does not provide any discrimination in ML algorithm.

```python
data= data.drop(columns = "Sensor6", axis=1)
```

### 3. Handling numerical feature with (likely) MCAR missing values

```python
data.info()
```

```python

column_list=[]
for column in data.columns:
    if '?' in data[column].values:
        column_list.append(column)
print(column_list)

```

```python
temp_df_categorical= pd.DataFrame(data=data[column_list[0]])
temp_df_numerical=pd.DataFrame(data=data[column_list[1]])
temp_df_numerical = pd.concat([temp_df_numerical.reset_index(drop=True), data[column_list[2]].reset_index(drop=True)], axis=1)
```

On `'temp_df_numerical'` we will apply Simple Imputer from sklearn to impute missing values. 

```python
temp_df_numerical
```

```python
from sklearn.impute import SimpleImputer
temp_df_numerical.replace('?', '', inplace=True)
imputer = SimpleImputer(missing_values='', strategy='most_frequent')
temp_df_new = pd.DataFrame(imputer.fit_transform(temp_df_numerical), columns=temp_df_numerical.columns)
print("Temp_df : \n")
print(temp_df_new)
```

### 4. One-hot encoding and handling categorical feature with (likely) MCAR missing value

On `'temp_df_categorical'` we will apply One Hot Encoding because **Mode** is a categorical feature. But before that we will also have to impute the missing values using Simple Imputer. 

```python
temp_df_categorical
```

```python
from sklearn.preprocessing import OneHotEncoder
temp_df_categorical.replace('?', '', inplace=True)
imputer = SimpleImputer(missing_values='', strategy='most_frequent')
temp_df_categorical= pd.DataFrame(imputer.fit_transform(temp_df_categorical), columns=temp_df_categorical.columns)
encode = OneHotEncoder(handle_unknown='ignore')

encoded = encode.fit_transform(temp_df_categorical[['Mode']])
encoded_data = pd.DataFrame(encoded.toarray(), columns=encode.get_feature_names_out(['Mode']))


data_encoded = pd.concat([temp_df_categorical, encoded_data], axis=1)


data_encoded.drop('Mode', axis=1, inplace=True)
print(data_encoded.head())
```

```python
temp_df_new= pd.concat([data_encoded.reset_index(drop=True), temp_df_new.reset_index(drop=True)], axis=1)
temp_df_new.head()
```

```python
temp_df_new.columns=['Mode_1', 'Mode_2', 'Mode_3', 'Sensor_3', 'Sensor_9']
```

```python
df_new = pd.concat([data.reset_index(drop=True), temp_df_new.reset_index(drop=True)], axis=1)
```

```python
df_new.head()
```

```python
df_new=df_new.drop(['Mode', 'Sensor3','Sensor9'], axis=1)
```

```python
df_new.head()
```

### 5. Standardization

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
numerical_features= ['Sensor1','Sensor_3', 'Sensor4', 'Sensor7','Sensor_9']
def standardization():
    num_pipeline= Pipeline(steps=
    [("scaler", StandardScaler())]
                )
           
    preprocessor=ColumnTransformer(
                [
                    ("num_pipeline", num_pipeline, numerical_features)
                ]
            )
    return preprocessor
```

```python
preprocessor = standardization()
dataframe_processed = preprocessor.fit_transform(dataframe)

```

```python
dataframe_processed
```

```python
dataframe_processed_df = pd.DataFrame(dataframe_processed, columns=numerical_features)
dataframe_processed_df.columns=['Sensor1_new', 'Sensor_3_new', 'Sensor4_new', 'Sensor7_new','Sensor_9_new']
dataframe_concatenated = pd.concat([dataframe, dataframe_processed_df], axis=1)
dataframe_concatenated.head()
```

```python
dataframe_dropped = dataframe_concatenated.drop(dataframe_concatenated.columns[1:4], axis=1)
print(dataframe_dropped)
```
